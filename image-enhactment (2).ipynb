```python
import cv2
import matplotlib.pyplot as plt


# Read an image
img_bgr = cv2.imread('../input/donutpink/donut_circle.png', 1)
plt.imshow(img_bgr)
plt.show()

# Histogram plotting of the image
color = ('b', 'g', 'r')

for i, col in enumerate(color):
	
	histr = cv2.calcHist([img_bgr],
						[i], None,
						[256],
						[0, 256])
	
	plt.plot(histr, color = col)
	
	# Limit X - axis to 256
	plt.xlim([0, 256])
	
plt.show()

# get height and width of the image
height, width, _ = img_bgr.shape

for i in range(0, height - 1):
	for j in range(0, width - 1):
		
		# Get the pixel value
		pixel = img_bgr[i, j]
		
		# Negate each channel by
		# subtracting it from 255
		
		# 1st index contains red pixel
		pixel[0] = 255 - pixel[0]
		
		# 2nd index contains green pixel
		pixel[1] = 255 - pixel[1]
		
		# 3rd index contains blue pixel
		pixel[2] = 255 - pixel[2]
		
		# Store new values in the pixel
		img_bgr[i, j] = pixel

# Display the negative transformed image
plt.imshow(img_bgr)
plt.show()

# Histogram plotting of the
# negative transformed image
color = ('b', 'g', 'r')

for i, col in enumerate(color):
	
	histr = cv2.calcHist([img_bgr],
						[i], None,
						[256],
						[0, 256])
	
	plt.plot(histr, color = col)
	plt.xlim([0, 256])
	
plt.show()
```

```python
#important library to show the image 
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
#importing numpy to work with large set of data.
import numpy as np
#image read function
img=mpimg.imread('../input/donutpink/donut_circle.png')
#image sclicing into 2D. 
x=img[:,:,0]
# x co-ordinate denotation. 
plt.xlabel("Value")
# y co-ordinate denotation.
plt.ylabel("pixels Frequency")
# title of an image .
plt.title("Original Image")
# imshow function with comperision of gray level value.
plt.imshow(x,cmap="gray")
#plot the image on a plane.
plt.show()

y=np.shape(x)
z=np.zeros(y)
#convert the image into its negative value.
z=255-x
plt.xlabel("Value")
plt.ylabel("pixels Frequency")
plt.title("Negative image ")
plt.imshow(z,cmap="gray")
plt.show()
```

```python
from PIL import Image, ImageEnhance
#read the image
im = Image.open("../input/icecream/icecream_circle.png")
#image brightness enhancer
enhancer = ImageEnhance.Brightness(im)
factor = 1 #gives original image
im_output = enhancer.enhance(factor)
im_output.save('asli.png')
factor = 0.5 #darkens the image
im_output = enhancer.enhance(factor)
im_output.save('gelap.png')
factor = 2.5 #brightens the image
im_output = enhancer.enhance(factor)
im_output.save('cerah.png')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
lego_a = cv2.imread('../input/donutpink/donut_circle.png',1)
lego_a = cv2.cvtColor(lego_a, cv2.COLOR_BGR2RGB)
lego_b = cv2.imread('../input/icecream/icecream_circle.png',1)
lego_b = cv2.cvtColor(lego_b, cv2.COLOR_BGR2RGB)
lego_blending = cv2.addWeighted(lego_a, 0.7, lego_b, 0.5,0)
plt.figure(figsize=(20, 20))
plt.subplot(1,3,1)
plt.imshow(lego_a)
plt.subplot(1,3,2)
plt.imshow(lego_b)
plt.subplot(1,3,3)
plt.imshow(lego_blending)
plt.show()
```

```python
img = cv2.imread('../input/donutpink/donut_circle.png',0)
gamma = 2.5
img_power = img**gamma
plt.figure(figsize=(20, 20))
plt.subplot(1,2,1)
plt.imshow(img,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_power,cmap="gray")
plt.show()
```

```python
img = cv2.imread('../input/icecream/icecream_circle.png',1)
kernel = np.ones((3, 3), np.float32) / 9
img_out = cv2.filter2D(img, -1, kernel)
plt.figure(figsize=(20, 20))
plt.subplot(1,2,2)
plt.imshow(img,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_out,cmap="gray")
plt.show()
```

```python
img = cv2.imread('../input/photos/photo_2022-06-19_17-06-48.jpg',0)
c = 255 / np.log10(1 + np.max(img))
img_log = c * (np.log10(img + 1))
img_log = np.array(img_log, dtype = np.uint8)
plt.figure(figsize=(20, 20))
plt.subplot(1,2,1)
plt.imshow(img,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_log,cmap="gray")
plt.show()
```
